# Matplotlib: Python’s Visualization Library

Customizable – Almost every aspect of a plot can be customized (e.g., labels, titles, colors).

Wide Range of Plot Types – Line, bar, scatter, histogram, heatmaps, and more.

Integrated with NumPy and Pandas – Great for data manipulation and visualization workflows.

Interactive Plots – Works with Jupyter Notebooks and can create interactive visualizations.

```python
import matplotlib.pyplot as plt
```

```python
# Example Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a simple line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Show the plot
plt.show()
```

```python
# Scatter plot: good for visualizing relationships between variables
plt.scatter(x, y, color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 2, 5, 9]

# Bar chart
plt.bar(categories, values, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
sizes = [25, 30, 45]
labels = ['Category A', 'Category B', 'Category C']

# Pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
# Create a figure with multiple subplots
fig, (ax1, ax2) = plt.subplots(1, 2)  # 1 row, 2 columns

# First subplot: Line plot
ax1.plot(x, y, color='blue')
ax1.set_title('Line Plot')

# Second subplot: Bar plot
ax2.bar(categories, values, color='orange')
ax2.set_title('Bar Chart')

plt.tight_layout()  # Adjust spacing between plots
plt.show()
```

```python
# Customizing the style of a plot
plt.plot(x, y, linestyle='--', color='red', marker='o')

# Adding grid
plt.grid(True)

# Adding labels and title with custom font size
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.title('Customized Line Plot', fontsize=16)

plt.show()
```
